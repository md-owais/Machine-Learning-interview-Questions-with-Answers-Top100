# 18. What are the assumptions of linear regression?

# Assumptions of Linear Regression

The model
Linear Regression works well only when certain assumptions are satisfied.

🔥 1. Linearity
📌 Meaning

The relationship between independent variables (X) and dependent variable (Y) should be linear.

🧠 Example
House price vs area → linear trend
❌ Violation
If relationship is curved → model performs poorly
🔥 2. Independence of Errors
📌 Meaning

Errors (residuals) should be independent of each other

🧠 Example
One data point error should not affect another
❌ Violation
Common in time series data
🔥 3. Homoscedasticity (Constant Variance)
📌 Meaning

The variance of errors should be constant across all values of X

🧠 Visual Idea
Spread of errors should be uniform
❌ Violation (Heteroscedasticity)
Errors increase/decrease with X
🔥 4. Normality of Errors
📌 Meaning

Residuals should follow a normal distribution

🧠 Why?
Important for statistical tests and confidence intervals
🔥 5. No Multicollinearity
📌 Meaning

Independent variables should not be highly correlated with each other

🧠 Example
height in cm and height in inches → highly correlated
❌ Problem
Makes coefficients unstable
🔥 6. No Autocorrelation
📌 Meaning

Residuals should not be correlated with each other over time

🧠 Example
Time series data often violates this
🔥 7. No Significant Outliers
📌 Meaning

Extreme values should not dominate the model

❌ Problem
Outliers can distort regression line
🔹 Summary Table
Assumption	Meaning
Linearity	X and Y have linear relationship
Independence	Errors are independent
Homoscedasticity	Constant error variance
Normality	Errors are normally distributed
No Multicollinearity	Features not highly correlated
No Autocorrelation	Errors not time-dependent
No Outliers	No extreme influence

🔹 Interview Answer (5 lines)

Linear regression assumes that there is a linear relationship between independent and dependent variables. The residuals should be independent and normally distributed with constant variance (homoscedasticity). There should be no multicollinearity among independent variables. Additionally, the model assumes no autocorrelation and no significant outliers. Violating these assumptions can lead to biased or inefficient estimates.
