```python
# Convex Function Example
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = x**2

plt.plot(x, y)
plt.title("Convex Function: y = x^2")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
# Non-Convex Function Example
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 500)
y = x**4 - 4*x**2 + x

plt.plot(x, y)
plt.title("Non-Convex Function")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
